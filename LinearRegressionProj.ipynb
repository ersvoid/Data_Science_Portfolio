```python
# SPLITTING DATA INTO TRAINING AND TESTING DATA
```

```python
customers.columns
```

```python
X = customers[['Avg. Session Length', 'Time on App',
       'Time on Website', 'Length of Membership']]
y = customers['Yearly Amount Spent']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
# TRAINING THE REGRESSION MODEL
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
lm.coef_
```

```python
# PREDICTING DATA
```

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test, predictions)
```

```python
# EVALUATING MODEL
```

```python
from sklearn import metrics
```

```python
print('MAE: ' + str(metrics.mean_absolute_error(y_test, predictions)))
print('MSE: ' + str(metrics.mean_squared_error(y_test, predictions)))
print('RMSE: ' + str(np.sqrt(metrics.mean_squared_error(y_test, predictions))))
```

```python
# PLOTTING RESIDUALS
```

```python
sns.displot((y_test - predictions), bins=50, kde=True)
```

```python
# CONCLUSION
```

```python
cdf = pd.DataFrame(lm.coef_, X.columns, columns=['Coeff'])
cdf
```

```python
# Each coefficient shows how much the amount spent by a customer will increase by an increase of the variable.
# This increase is only if all the other variables are held constant.
# The coefficients imply that there is a noticeable difference between customers using the app and using the website. 
# Customers who spend time on the app are more likely to spend more money than people on the website. 
# However, the greatest contribution to money spent, is how long they have had a membership.
```
